```python
import pandas as pd #Mengimpor library pandas sebagai pd untuk manipulasi data.
import matplotlib.pyplot as plt #Mengimpor library matplotlib.pyplot sebagai plt untuk pembuatan plot.
import numpy as np # Mengimpor library numpy sebagai np untuk operasi numerik.
22
# Membuat dictionary dari dataset
data = {
    "Variant": ["Matcha", "Vanilla", "Pistachio", "Karamel", "Cokelat Mint", "Cokelat", "Pisang"],
    "Flavor Rating": [2, 6, 4, 4, 7, 9, 1],
    "Texture Rating": [0, 9, 0, 5, 5, 8, 3],
    "Overall Rating": [3, 8, 4, 5, 6, 9, 4]
}

# Membuat DataFrame dari dictionary data
df = pd.DataFrame(data)

# Set index ke kolom 'Variant'
df = df.set_index('Variant')

# Membuat plot bar
plt.figure(figsize=(10, 6))  # Mengatur ukuran figure
df.plot(kind='bar')  # Membuat bar plot dari DataFrame
plt.title('Ice Cream Ratings')  # Menambahkan judul plot
plt.xlabel('Variant')  # Menambahkan label sumbu x
plt.ylabel('Ratings')  # Menambahkan label sumbu y
plt.xticks(rotation=45)  # Memutar label sumbu x sebanyak 45 derajat
plt.legend(loc='upper right')  # Menempatkan legenda/keterangan di sudut kanan atas
plt.show()  # Menampilkan plot

# Membuat heatmap menggunakan matplotlib
plt.figure(figsize=(10, 6))  # Mengatur ukuran figure
plt.imshow(df, cmap='coolwarm', aspect='auto')  # Menampilkan heatmap dari DataFrame dengan colormap coolwarm
plt.colorbar(label='Rating')  # Menambahkan colorbar dengan label 'Rating'
plt.xticks(np.arange(len(df.columns)), df.columns, rotation=45)  # Mengatur label sumbu x
plt.yticks(np.arange(len(df.index)), df.index)  # Mengatur label sumbu y
plt.title('Ice Cream Ratings Heatmap')  # Menambahkan judul heatmap

# Menambahkan anotasi pada heatmap
for i in range(len(df.index)): #Memulai loop untuk tiap baris dalam DataFrame.
    for j in range(len(df.columns)): #Memulai loop untuk tiap kolom dalam DataFrame.
        plt.text(j, i, df.iloc[i, j], ha='center', va='center', color='black')  # Menambahkan teks di tiap sel heatmap

plt.show()  # Menampilkan heatmap
```
